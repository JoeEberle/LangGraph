## Creating a Graph NODE 

```python
class AgentState(TypedDict):
    name:str
    age:int
    is_student: bool
```

```python
def ask_name(state: AgentState) -> AgentState:
    """ 
    Prompts the user to enter their name and updates the state with it. 
    """
    print("What's your name?") 
    name = input(">> ")
    state["name"] = name 
    return state
```

## Creating a Second Graph NODE 

```python
def greet(state: AgentState) -> AgentState:
    """
    Greets the user using name stored in the state. 
    """
    print(f"Hello, {state["name"]}! Welcome to LangGraph")
    return state 
```

```python
import random

def chat(state: AgentState) -> AgentState:
    """
    Makes a random chat conversation response about learning LangGraph
    """
    response = random.randint(1, 4)
    name = state["name"]

    if response == 1:
        print(f"Great job {name}, learning the basics of LangGraph!")
    elif response == 2:
        print(f"Good progress {name}, keep going!")
    elif response == 3:
        print(f"Nice job {name}  way to learn!!")
    else:
        print(f"{name}, your LangGraph workflow is getting better!!")

    return state
```

```python
graph = StateGraph(AgentState)
graph.add_node("ask_name", ask_name)
graph.add_node("greet", greet) 
graph.add_node("chat", chat) 

graph.set_entry_point("ask_name") 
```

## Adding Edges 

```python
graph.add_edge("ask_name","greet") 
graph.add_edge("greet","chat") 
graph.add_edge("chat",END) 
```

```python
compiled_graph = graph.compile()
```

```python
print(compiled_graph.get_graph().draw_ascii())
```

## Chapter 4 - Multiple Inputs Graph 

```python
from typing import TypedDict, List
from langgraph.graph import StateGraph, END 
```

```python
class AgentState(TypedDict):
    name:str
    values:List[int]
    result:str
```

```python
def ask_values(state: AgentState) -> AgentState:
    """ Prompts the user to enter a list of recent purchasa amounts """ 
    print(f"Lets try the budget analyzer.")     
    print(f"Hi {state['name']}! Enter your last 5 financial transactions seperate by commas:") 
    values = input(">> ")
    try:
        values = [int(v.strip()) for v in values.split(",") if v.strip()] 
        if len(values) < 1:
            raise ValueError()
    except ValueError:
        print("please on enter valid numbers seperated by commas.") 
        return ask_values(state)
    state["values"] = values
    return state
```

```python
def analyze_and_report(state: AgentState) -> AgentState:
    """ Analyzes the values and produces result """ 
    total = sum(state["values"]) 
    result = "Saver" if total > 0 else "Spender" 
    print(f"Total spent: ${total}") 
    print(f"Result - your a {result} !")          
    return state
```

```python
graph = StateGraph(AgentState)
graph.add_node("ask_name", ask_name)
graph.add_node("ask_values", ask_values) 
graph.add_node("analyze_and_report", analyze_and_report) 
graph.add_edge("ask_name","ask_values") 
graph.add_edge("ask_values","analyze_and_report") 
graph.add_edge("analyze_and_report",END) 
graph.set_entry_point("ask_name") 
```

```python
compiled_graph = graph.compile()
print(compiled_graph.get_graph().draw_ascii())
```

## Chapter 5 - Conditional Graphs 

```python
definition ='''
##  Conditional Graphs

1. **Define the goal **  Clearly state the single problem or capability the graph is meant to solve.
2. **Design the shared state**  Decide what information must persist and evolve across nodes.
3. **Identify the nodes**  Break the workflow into small, single-purpose steps.
4. **Implement node functions**  Write pure functions that take state in and return updated state.
5. **Determine conditional logic**  Identify where decisions, branching, or retries are required.
6. **Connect the edges**  Wire nodes together using static and conditional transitions.
7. **Set start   and end  points**  Explicitly define how the graph begins and terminates.
8. **Compile the graph**  Validate and lock the graph structure for execution.
9. **Test with simple inputs**  Run minimal cases to verify state transitions and routing.
10. **Add guardrails and logging**  Apply limits, error handling, and observability for safe execution.

'''

sb.outmd(definition)
```

```python
from langgraph.graph import StateGraph, END
from typing import TypedDict, Literal 

class TipCalcState(TypedDict):
    bill : float
    adjustAmount: float
    operation: Literal["+","-"]
    final_total: float
```

```python
def get_user_input(state: TipCalcState) -> TipCalcState:
    print(f"Enter your bill amount:")
    state["bill"] = float(input(">> ")) 

    print(f"Do you want to add a tip (+) or a discount (-) :")
    op = input(">>").strip()
    print(f" the operation sellected is:{op} ") 
    if op not in {"+","-"}:
        print(" Invalid operation, defaulting to tip (+)")
        op = "+" 
    state["operation"] = op  

    operation_name = "tip" if op == "+" else "discount" 
    print(f"Enter amount of {operation_name}:")
    state["adjustAmount"] = float(input(">> ")) 
    return state 
```

```python
def apply_tip(state: TipCalcState) -> TipCalcState:
    state["finalTotal"] = state["bill"] + state["adjustAmount"]  
    print(f" Final amount after tip: {state['finalTotal']} ")
    return state
```

```python
def apply_discount(state: TipCalcState) -> TipCalcState:
    state["finalTotal"] = state["bill"] - state["adjustAmount"]  
    print(f" Final amount after discount: {state['finalTotal']} ")
    return state
```

```python
def get_operation(state):
    return state["operation"]
```

## define the conditional Graph  

```python
builder = StateGraph(TipCalcState)

builder.add_node("input", get_user_input)
builder.add_node("tip", apply_tip)
builder.add_node("discount", apply_discount)

builder.set_entry_point("input")

builder.add_conditional_edges(
    "input",
    get_operation,
    {
        "+": "tip",
        "-": "discount" 
    }
)

builder.add_edge("tip", END)
builder.add_edge("discount", END) 

graph =builder.compile()
```

```python
testing = False
if testing:
    graph.invoke({"bill":0,"adjustamount":0,"operation":"+","final_total":0})
```

## Excercise 5 chapter 5 Optional Rounding - Conditional Graphs

```python
from langgraph.graph import StateGraph, END
from typing import TypedDict, Literal
```

```python
class TipClassState(TypedDict):
    bill:float
    adjustAmount:float
    operation: Literal ["+","-"]
    final_total:float
    round: Literal["yes","no"]    
```

```python
def get_user_input(state: TipClassState) -> TipClassState:
    print("1 enter your total bill amount: ")
    state["bill"] = float(input(">> "))

    print("2 Do you want to add tip (+) or apply a discount (-) ")  
    op = input(">> ").strip() 
    if op not in ["+","-"]:
        print("Invalid operation applying tip (+) as default  ")
        op = "+"
    state["operation"] = op 

    print("3 Enter the tip or discount amount ")  
    state["adjustAmount"] = float(input(">> "))
    
    return state
```

```python
def apply_tip(state: TipCalcState) -> TipCalcState:
    state["final_total"] = state["bill"] + state["adjustAmount"]  
    print(f" Final amount after tip: {state['final_total']:.2f} ")
    return state
```

```python
def apply_discount(state: TipCalcState) -> TipCalcState:
    state["final_total"] = state["bill"] - state["adjustAmount"]  
    print(f" Final amount after discount: {state['final_total']:.2f} ")
    return state
```

```python
def ask_rounding(state: TipCalcState) -> TipCalcState:
    print("4 Would you like to round to nearest dollar (yes/no) ")
    round_choice = input(">>: ").strip().lower()
    if round_choice in ["yes","no"]:
        state["round"] = "yes" if round_choice == "yes" else "no"
    return(state)
```

```python
def round_total(state: TipCalcState) -> TipCalcState:
    state["final_total"] = round(state["final_total"])
    print(f"Final total after rounding: ${state["final_total"]:.2f} ") 
    return state                                 
```

```python
def skip_rounding(state: TipCalcState) -> TipCalcState:
    print(f"Final total no rounding: ${state["final_total"]:.2f} ") 
    return state  
```

```python
# Build the graph 
builder = StateGraph(TipCalcState)

builder.add_node("input", get_user_input) 
builder.add_node("tip",apply_tip)
builder.add_node("discount",apply_discount)
builder.add_node("ask_rounding",ask_rounding)
builder.add_node("round_total",round_total)
builder.add_node("skip_rounding",skip_rounding)

builder.set_entry_point("input") 

# first conditional choice 
builder.add_conditional_edges(
    "input",
    lambda state: state["operation"],
    {
        "+": "tip",
        "-": "discount"
    }
)

# after tip ask about rounding 
builder.add_edge("tip", "ask_rounding") 
builder.add_edge("discount", "ask_rounding") 

# after rounding, conditionally round or skip 
builder.add_conditional_edges(
    "ask_rounding",
    lambda state: state["round"],
    {
        "yes": "round_total",
        "no": "skip_rounding"
    }
)

# add END
builder.add_edge("round_total", END) 
builder.add_edge("skip_rounding", END) 
```

```python
compiled_graph = builder.compile()
print(compiled_graph.get_graph().draw_ascii())
```

```python
testing = True
if testing:
    # run the graph
    graph = builder.compile()
    graph.invoke({
        "bill": 0,
        "adjustAmount": 0, 
        "operation": "+", 
        "final_total" : 0,    
        "round" : "yes"
    }) 
```

## Chapter 6 looping Graph

```python
from typing import TypedDict
from langgraph.graph import StateGraph, END

class AuthState(TypedDict):
    attempt: int
    authentication: bool

CORRECT_PASSWORD = "secure1234"

def get_password(state: AuthState) -> AuthState:
    print(f'attempt {state["attempt"] + 1} - enter your password: ')
    pw = input(">> ").strip()
    state["attempt"] += 1
    state["authentication"] = (pw == CORRECT_PASSWORD)
    return state


def success(state: AuthState) -> AuthState:
    print(" Authentication granted!")
    return state


def check_auth(state: AuthState) -> str:
    return "success" if state["authentication"] else "retry"

builder = StateGraph(AuthState)
builder.add_node("get_password", get_password)
builder.add_node("success", success)

builder.set_entry_point("get_password")

builder.add_conditional_edges(
    "get_password",
    check_auth,
    {
        "success": "success",       # proceed
        "retry": "get_password"     # loop back
    }
)

builder.add_edge("success", END)

graph = builder.compile()

# run it
result = graph.invoke({"attempt": 0, "authentication": False})
print(result)
```
